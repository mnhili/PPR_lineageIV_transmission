```python
import numpy as np
import matplotlib.pyplot as plt

# Initialisation
q = np.full(100, np.nan)
pp = np.linspace(0.0001, 0.1, 100)
N = 20

# Calcul de la probabilité de détection
for i in range(100):
    # Génération de 1000 réalisations de binomiale avec des probabilités pp[i]
    q[i] = np.sum(np.random.binomial(N, pp[i], 1000) > 0) / 1000

# Création de la figure
fig = plt.figure()

# Plotting avec des cercles non remplis
plt.plot(pp, q, 'o', markerfacecolor='none')  # 'o' pour des cercles non remplis
plt.xlabel("Probability of Transmission")
plt.ylabel("Probability of Detecting")

# Ajuster l'axe y de 0 à 1, par pas de 0.1
plt.yticks(np.arange(0, 1.1, 0.1))

# Ajouter les lignes horizontales et verticales
plt.axvline(x=0.05, color='red', linestyle='--')
plt.axhline(y=0.8, color='red', linestyle='--')

# Sauvegarde de la figure
fig.savefig('powerAnalysis_above_24h.tiff', dpi=350, format='tiff')

# Affichage du graphique
plt.show()
```
