```python
import pandas as pd

df = pd.read_excel("experience_1_1_1h.xlsx")
df
```

```python
df.dtypes
```

```python
# Encode Cap1 column as a category 
df['Cap1'] = df['Cap1'].astype('category')

# Encode Cap2 column as a category 
df['Cap2'] = df['Cap2'].astype('category')
df.dtypes
```

```python
# Keep only rows where either Cap1 or Cap2 is the seed (infected)
df = df[(df['Cap1'] == 436) | (df['Cap2'] == 436)]

df
```

```python
df.describe()
```

```python
df
```

```python
# Convert time from seconds to minutes  
df.loc[:, 'Time'] = df['Time'] / 60 

df.loc[:, 'Time'] = df.loc[:, 'Time'].round(0)

df.head(50)
```

```python
# Group by 'Patient 1', 'Patient 2', and 'Time', aggregating 'Distance' by taking the mean
df = df.groupby(['Cap1', 'Cap2', 'Time'])['Distance'].min().reset_index()
```

```python
df
```

```python
# Save the cleaned data to a new CSV file
df.to_csv("cleaned_experiment_1_1_1h.csv", index=False)
```
