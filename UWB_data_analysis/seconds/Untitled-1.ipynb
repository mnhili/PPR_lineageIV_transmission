```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 5))

for i, df in enumerate([df1, df2, df3]):
    data = [df[df['Cap2'] == individu]['Distance'] for individu in df['Cap2'].unique()]
    axes[i].boxplot(data)
    axes[i].set_xticklabels(df['Cap2'].unique(), rotation=45)
    axes[i].set_title(f'Fichier {i+1}')
    axes[i].set_ylabel('Distance')

plt.tight_layout()
plt.show()
```

```python
# Create a figure with subplots for each Excel file
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 5))

# Plot boxplots for each file
for i, file_df in enumerate([df1, df2, df3]):
    ax = axes[i]
    ax.boxplot([file_df[file_df['Cap2'] == individu]['Distance']
                for individu in file_df['Cap2'].unique()],
               labels=file_df['Cap2'].unique(),
               vert=False)
    ax.set_title(f'File {i+1}')
    ax.set_xlabel('Distance')

# Adjust spacing and display the plot
plt.tight_layout()
plt.show()
```

```python
data = pd.concat([df1.assign(file='file1'), 
                  df2.assign(file='file2'),
                  df3.assign(file='file3')], ignore_index=True)

# Create a FacetGrid with the 'file' column as the row
g = sns.FacetGrid(data, row='file', height=4, aspect=3)

# Draw a boxplot for each individual's heart rate/sec
g.map_dataframe(sns.boxplot, x='Cap2', y='Distance', 
                hue='Cap2', flierprops={'markerfacecolor': 'red', 'marker': 'o'},
                legend=False)

# Adjust the axis labels and titles
g.set_axis_labels('Cap2', 'Distance')
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('Boxplots for Heart Rate/sec by Individual and File')

# Display the plot
plt.show()
```
